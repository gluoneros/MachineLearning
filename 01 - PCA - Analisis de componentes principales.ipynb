```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x = 3*np.random.rand(200)# generate 200 random numbers between 0 and 3
y = 20*x + 2*np.random.randn(200)

x = x.reshape(200,1) #create a 2D array with one column
y = y.reshape(200,1)

print(x)
print(y)
```

```python
xy = np.hstack([x, y])# combina x and y into a single 2D array
print(xy.shape)
```

```python
plt.plot(xy[:,0], xy[:,1], '.')
plt.show()
```

```python
xy_centrado = xy - np.mean(xy, axis = 0)
```

```python
plt.plot(xy_centrado[:,0], xy_centrado[:,1], '.')
plt.show()
```

```python
autovalores, autovectores = np.linalg.eig(xy_centrado.T.dot(xy_centrado))# calcula autovslores and autovectores
print(autovectores)
```

```python
def graficarVectores(vecs, cols, alpha = 1):
    
    plt.axvline(x = 0, color = "grey", zorder = 0)
    plt.axhline(y = 0, color = "grey", zorder = 0)
    
    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                    [x[1]],
                    [x[2]],
                    [x[3]],
                    angles = 'xy', scale_units = 'xy',
                    scale = 1,
                    color = cols[i],
                    alpha = alpha
                    )
```

```python
graficarVectores(autovectores.T, ['blue', 'red'])

plt.plot(xy_centrado[:, 0], xy_centrado[:, 1]/20, '.')

plt.show()
```
